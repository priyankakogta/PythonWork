```python
import pandas as pd

data = pd.read_csv('amazon-orders.csv')
data.head(59)
#data.shape # This gives us the dimensions of the file, so there are 59 rows and 23 columns.
data = data.fillna(0)
data.head()
# data = data["Total Charged"].str.replace('$','') #This gave just the price for each row.
data["Total Charged"] = data["Total Charged"].str.replace('$','').astype(float) # Need to change to float or else it processes the values as strings.
data.head(59)
data["Total Charged"].sum() # $1777.73
```

```python
import pandas as pd

data = pd.read_csv('amazon-orders.csv')
data.head(59)
#data.shape # This gives us the dimensions of the file, so there are 59 rows and 23 columns.
data = data.fillna(0)
data.head()
# data = data["Total Charged"].str.replace('$','') #This gave just the price for each row.
data["Total Charged"] = data["Total Charged"].str.replace('$','').astype(float) # Need to change to float or else it processes the values as strings.
data.head(59)
data["Total Charged"].sum() # $1777.73
data["Total Charged"].min() # $ 1.04
```
